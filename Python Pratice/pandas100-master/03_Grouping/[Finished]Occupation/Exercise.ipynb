### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by="occupation").age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.head()
```

```python
a = users[users.gender == "M"].groupby(by="occupation")["gender"].count()/ \
users.groupby(by="occupation")["gender"].count()
a.sort_values(ascending=False)*100
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(by="occupation")["age"].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by=["occupation","gender"])["age"].agg(['mean'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
def male(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
    
def female(x):
    if x == 'M':
        return 0
    if x == 'F':
        return 1
    
def all(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 1
    
users["female"] = users.gender.apply(female)
users["male"] = users.gender.apply(male)
users["all"] = users.gender.apply(all)
```

```python
users.head()
```

```python
temp = users.groupby("occupation")[["female","male","all"]].sum()
temp.female = temp.female/temp['all']
temp.male = temp.male/temp['all']
```
